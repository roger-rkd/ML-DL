```python
import nltk
from nltk.stem import PorterStemmer # library used for stemming (finding the base root word)
from nltk.corpus import stopwords # library used for stopwords
```

```python
# TOKENIZATION 

# conversion of paragraph >> sentences >> words
```

```python
nltk.download('punkt_tab') # dependency to use sent_tokenize
sentences=nltk.sent_tokenize(paragraph) # this line would convert the paragraph into sentence
```

```python
print(sentences)
```

```python
type(sentences)
```

```python
stemmer=PorterStemmer() # implementation of stemming
```

```python
stemmer.stem('history')
```

```python
nltk.download('wordnet') # library used for lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('history') # output is the meaningful word
```

### STEP 1: Clean the paragraph

```python
len(sentences)
```

```python
import re # importing regular expression to clean the text
corpus=[]
for i in range(len(sentences)): # looping through the total number of sentences
    review=re.sub('[^a-zA-Z]',' ', sentences[i]) # substituting characters other than a-z and A-Z with a blank space for the variable 'sentences'
    review=review.lower() # converting the list into lowercase
    corpus.append(review) # appending them in the new list variable called corpus
```

```python
corpus
```

### STEP 2: Apply Stemming

```python
import nltk
nltk.download('stopwords') # downloading stopwords from nltk
for i in corpus: # looping through the entire corpus
    words=nltk.word_tokenize(i) # conversion of sentences into words
    for word in words: # taking one word at a time
        if word not in set(stopwords.words('english')): # checking if the word is in the stopword list of english language
            print(stemmer.stem(word)) # stemming on each word
```

### STEP 3: Apply Lemmatization

```python
import nltk
nltk.download('stopwords') # downloading stopwords from nltk
for i in corpus: # looping through the entire corpus
    words=nltk.word_tokenize(i) # conversion of sentences into words
    for word in words: # taking one word at a time
        if word not in set(stopwords.words('english')): # checking if the word is in the stopword list of english language
            print(lemmatizer.lemmatize(word)) # lemmatizing on each word
```
